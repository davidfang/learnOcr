```python
import pytesseract
from PIL import Image
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('./images/lena.png')
```

```python
px = img[100,90]
print(px)
```

```python
# 只获取蓝色 blue 通道的值
px_blue = img[100,90,0]
print(px_blue)
```

```python
img[100,90] = [255,255,255]
print(img[100,90])
```

```python
#图片属性
print(img.shape)
# 形状中包括行数、列数、通道
height,width,channels = img.shape
```

```python
#img.dtype获取图像数据类型：
print(img.dtype)
```

```python
#img.size获取图像总像素数：
print(img.size)#512*512*3
```

```python
# 截取脸部 ROI
face = img[100:200,115:188]
plt.imshow(face[:,:,::-1])
```

```python
# 通道分割与合并
b, g, r = cv2.split(img)
img2 = cv2.merge((b,g,r))
```

```python
#split()函数比较耗时，更高效的方式是用 numpy 中的索引，如提取 B 通道：
b = img[:,:,0]
plt.imshow(b)
```
