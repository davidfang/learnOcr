##### [05: 颜色空间转换](https://codec.wang/docs/opencv/start/changing-colorspaces)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 颜色空间转换
img = cv2.imread('./images/lena.png')

# 转换为灰度图
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
fig, axs = plt.subplots(2,2,figsize=(8,8))
axs[0,0].imshow(img)
axs[0,0].set_title('img')
axs[1,0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
axs[0,1].imshow(img_gray)
axs[0,1].set_title('img_gray')
axs[1,1].imshow(cv2.cvtColor(img_gray,cv2.COLOR_GRAY2RGB))
```

cv2.cvtColor()用来进行颜色模型转换，参数 1 是要转换的图片，参数 2 是转换模式， COLOR_BGR2GRAY表示 BGR→Gray，可用下面的代码显示所有的转换模式：

```python
flags = [i for i in dir(cv2) if i.startswith('COLOR_')]
flags
```

```python
# 蓝色的范围，不同光照条件下不一样，可灵活调整
lower_blue = np.array([100, 110, 110])
upper_blue = np.array([130, 255, 255])

## 2.从 BGR 转换到 HSV
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
# 3.inRange()：介于 lower/upper 之间的为白色，其余黑色
mask = cv2.inRange(hsv, lower_blue, upper_blue)
# 4.只保留原图中的蓝色部分
res = cv2.bitwise_and(img, img, mask=mask)
```

```python
fig, axs = plt.subplots(2,2,figsize=(8,8))
axs[0,0].imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
axs[0,0].set_title('img')
axs[0,1].imshow(cv2.cvtColor(hsv,cv2.COLOR_BGR2RGB))
axs[0,1].set_title('hsv')
axs[1,0].imshow(mask)
axs[1,0].set_title('mask')
axs[1,1].imshow(res)
axs[1,1].set_title('res')
```
