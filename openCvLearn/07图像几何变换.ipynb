```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
# 设置中文字体
# font = FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc', size=12)
```

```python
img = cv2.imread('./images/drawing.jpg')
# 按照指定的宽度、高度缩放图片
res = cv2.resize(img,(132,150))
# 按照比例缩放，如 x,y 轴均放大一倍
res2 = cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_LINEAR)

plt.figure(figsize=(12,8))

plt.subplot(1,3,1)
plt.imshow(img[:,:,::-1])
plt.subplot(1,3,2)
plt.imshow(res[:,:,::-1])
plt.subplot(1,3,3)
plt.imshow(res2[:,:,::-1])
```

```python
plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei']  # 使用文泉驿字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题
```

```python
img2 = cv2.imread('./images/lena.png')
flip1 = cv2.flip(img2,1)
flip0 = cv2.flip(img2,0)
flip2 = cv2.flip(img2,1)
flip3 = cv2.flip(img2,-1)
titles = ['原图','水平翻转','垂直翻转','水平垂直翻转']
images = [img2,flip1,flip0,flip3]

for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i][:,:,::-1])
    plt.title(titles[i])

plt.show()
```
