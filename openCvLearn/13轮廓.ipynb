###### [13: 轮廓](https://codec.wang/docs/opencv/basic/contours)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei']  # 使用文泉驿字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题
```

```python
img = cv2.imread('./images/handwriting.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# 寻找二值化图中的轮廓
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(len(contours))  # 结果应该为 2
```

```python
edges = cv2.Canny(img,30,70) # canny 边缘检测
edges2 = cv2.drawContours(img,contours,-1,(0,0,255),2)

plt.figure(figsize=(12,8))
plt.subplot(131),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.subplot(132),plt.imshow(cv2.cvtColor(edges,cv2.COLOR_BGR2RGB))
plt.subplot(133),plt.imshow(cv2.cvtColor(edges2,cv2.COLOR_BGR2RGB))
```
