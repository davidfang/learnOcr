##### [仿射变换与透视变换](https://codec.wang/docs/opencv/start/extra-05-warpaffine-warpperspective)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.sans-serif'] = ['WenQuanYi Zen Hei']  # 使用文泉驿字体
plt.rcParams['axes.unicode_minus'] = False  # 解决坐标轴负号显示问题
```

```python
img = cv2.imread('./images/card.jpg')
# 原图中卡片的四个角点
pts1 = np.float32([[148, 80], [437, 114], [94, 247], [423, 288]])
# 变换后分别在左上、右上、左下、右下四个点
pts2 = np.float32([[0, 0], [320, 0], [0, 178], [320, 178]])

# 生成透视变换矩阵
M = cv2.getPerspectiveTransform(pts1,pts2)
# 进行透视变换，参数 3 是目标图像大小
dst = cv2.warpPerspective(img, M, (320,178))

plt.subplot(121),plt.imshow(img[:,:,::-1]),plt.title('输入')
plt.subplot(122),plt.imshow(dst[:,:,::-1]),plt.title('输出')

plt.show()
```
